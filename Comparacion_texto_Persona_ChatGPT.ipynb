####Clases,objetos y metodos

En programación orientada a objetos, una clase es una plantilla o un plano que define la estructura y el comportamiento de un objeto. Una clase describe las propiedades (atributos) que tiene un objeto y los comportamientos (métodos) que puede realizar.

Por ejemplo, si tienes una clase llamada "Perro", puedes definir los atributos de un perro, como su nombre, edad y raza, así como los métodos asociados, como ladrar, comer y dormir. La clase "Perro" es solo una plantilla que define cómo se verán y se comportarán los objetos de tipo "Perro".

Por otro lado, un objeto es una instancia particular de una clase. Cuando creas un objeto, estás creando una copia de la estructura definida por la clase. Cada objeto tiene su propio estado (valores de los atributos) y puede realizar las acciones definidas por los métodos de su clase.

Siguiendo el ejemplo anterior, si creas un objeto de la clase "Perro" llamado "miPerro", estarás creando una instancia individual con un nombre, una edad y una raza específicos. Puedes llamar a los métodos de "miPerro", como hacerlo ladrar o alimentarlo.
