## Fraccion de Stopwords en un corpus

En el procesamiento del lenguaje natural (PLN), los "stopwords" son palabras que se consideran frecuentes y comunes en un idioma, pero que a menudo no aportan un significado sustancial al contexto del texto. Estas palabras suelen ser artículos, pronombres, preposiciones y otros términos que se utilizan con frecuencia en la construcción de oraciones, pero que en sí mismos no transmiten información relevante sobre el tema del texto.

La eliminación de stopwords es una técnica común en el preprocesamiento de texto en PLN. Al eliminar estas palabras de un texto, se puede reducir el ruido y el tamaño del vocabulario, lo que puede ayudar a mejorar la eficiencia y la calidad de los análisis posteriores, como la clasificación de texto, la agrupación de documentos o la extracción de características.

```
def stopwords_percentage(text):
  '''
  aqui usamos un recurso léxico (stopwords) para filtrar un corpus
  '''
  stopwd = stopwords.words('english')
  content = [w  for w in text if w.lower()  in stopwd]  #w.lower pasa todo a minuscula.
  return len(content)/len(text)

stopwords_percentage(text1)
```

# Explorando los Synsets

Para explorar cada synset y acceder a la información asociada, puedes utilizar los métodos y atributos proporcionados por NLTK y WordNet. Aquí hay algunos ejemplos de cómo puedes explorar cada synset en la lista ss:



1.  Acceder a la definición de un synset:


```
synset = ss[0]  # Por ejemplo, el primer synset en la lista(car.n.01)
definition = synset.definition()
print("Definición:", definition)
```

2.  Obtener los ejemplos de uso de un synset:



```
synset = ss[2]  # Por ejemplo, el tercer synset en la lista(carrier.n.02)
examples = synset.examples()   #Aquí se está llamando al método examples() del objeto synset. Este método devuelve una lista de ejemplos de uso asociados con el synset.
                               #Cada ejemplo es una cadena de texto que ilustra cómo se usa la palabra representada por el synset en el contexto.
print("Ejemplos de uso:")
for example in examples:       #Aquí comienza un bucle for que recorre cada elemento de la lista examples.
    print("-", example)        #Se imprime cada ejemplo de uso: en este caso solo hay uno
```



3.   Obtener los sinónimos (lemmas) de un synset:



```
synset = ss[0]                                          #Aquí se asigna a la variable synset el primer elemento de la lista ss.
synonyms = [lemma.name() for lemma in synset.lemmas()]  #Se crea una lista llamada synonyms utilizando una comprensión de lista.
                                                        #Para cada objeto lemma en el atributo lemmas() del synset, se llama al método name() del objeto lemma para obtener el nombre del sinónimo.
print("Sinónimos:", synonyms)
```



4.   Explorar las relaciones semánticas de un synset:



```
synset = ss[0]
hypernyms = synset.hypernyms()  # Obtener hiperónimos (palabras más generales).Se llama al método hypernyms() del objeto synset.
                                # Este método devuelve una lista de hiperónimos, es decir, palabras más generales que engloban el concepto representado por el synset. "hypernyms" es una lista
hyponyms = synset.hyponyms()    # Obtener hipónimos (palabras más específicas). Se llama al método hyponyms() del objeto synset.
                                #Este método devuelve una lista de hipónimos, es decir, palabras más específicas que son subordinadas o ejemplos del concepto representado por el synset.

print("Hiperónimos:")
for hypernym in hypernyms:        #bucle for que recorre cada elemento de la lista hypernyms
    print("-", hypernym.name())   #Imprime los elementos de hypernyms precedidos por un guion

print("Hipónimos:")
for hyponym in hyponyms:
    print("-", hyponym.name())
```

### Utilizamos un bucle para explorar los distintos synsets de la palabra carro

```
# synset: grupo de sinómimos de una palabra.
ss = wn.synsets('carro', lang='spa')

# explorando los synsets
for syn in ss:
  print(syn.name(), ': ', syn.definition()) #Para cada synset en la lista `ss`, se imprime su nombre utilizando el método `name()` y su definición utilizando el método `definition()`.
  for name in syn.lemma_names():            #Para cada synset en la lista `ss`, se recorre cada uno de sus sinónimos (lemmas) utilizando el método `lemma_names()`.
    print(' * ', name)
```

```
# visualization references
# * http://www.randomhacks.net/2009/12/29/visualizing-wordnet-relationships-as-graphs/
# * http://dlacombejr.github.io/programming/2015/09/28/visualizing-cifar-10-categories-with-wordnet-and-networkx.html

import networkx as nx
import matplotlib.pyplot as plt

def closure_graph(synset, fn):
    seen = set()
    graph = nx.DiGraph()
    labels = {}

    def recurse(s):
        if not s in seen:
            seen.add(s)
            labels[s.name] = s.name().split('.')[0]
            graph.add_node(s.name)
            for s1 in fn(s):
                graph.add_node(s1.name)
                graph.add_edge(s.name, s1.name)
                recurse(s1)

    recurse(synset)
    return graph, labels

def draw_text_graph(G, labels):
    plt.figure(figsize=(18,12))
    pos = nx.planar_layout(G, scale=18)
    nx.draw_networkx_nodes(G, pos, node_color="red", linewidths=0, node_size=500)
    nx.draw_networkx_labels(G, pos, font_size=20, labels=labels)
    nx.draw_networkx_edges(G, pos)
    plt.xticks([])
    plt.yticks([])
```

## **Hyponyms:** Conceptos que son más especificos que la palabra raiz de la cual derivan.

```
print(ss[0].name())
G, labels = closure_graph(ss[0], fn = lambda s: s.hyponyms())
draw_text_graph(G, labels)
```

## **Hypernyms**: conceptos que son mas generales !

```
print(ss[0].name())
G, labels = closure_graph(ss[0], fn = lambda s: s.hypernyms())
draw_text_graph(G, labels)
```

## Similitud Semántica

Armamos una funcion que muestre nombre y definicion del synset y los lemas



```
def show_syns(word):
  ss = wn.synsets(word, lang='spa')
  for syn in ss:
    print(syn.name(), ': ', syn.definition())
    for name in syn.lemma_names():
      print(' * ', name)
  return ss
```

```
ss = show_syns('perro')
```

```
ss2 = show_syns('gato')
```

```
import matplotlib.pyplot as plt
import networkx as nx

def traverse(graph, start, node):
  graph.depth[node.name] = node.shortest_path_distance(start)
  for child in node.hyponyms():
    graph.add_edge(node.name, child.name)
    traverse(graph, start, child)

def hyponym_graph(start):
  G = nx.Graph()
  G.depth = {}
  traverse(G, start, start)
  return G

def graph_draw(graph):
  node_sizes = [16 * graph.degree(n) for n in graph.nodes()]
  node_colors = [graph.depth[n] for n in graph.nodes()]

  pos = nx.spring_layout(graph)  # Layout del grafo

  nx.draw(graph, pos,
          node_size=node_sizes,
          node_color=node_colors,
          with_labels=True)
  plt.show()
```

```
dog = wn.synset('dog.n.01')
graph = hyponym_graph(dog)
graph_draw(graph)
```
